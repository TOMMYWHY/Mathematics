```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5,101)

def math_function(x,y):
    plt.figure(figsize=(6,3),dpi=100)
    plt.plot(x,y,color="m")
    plt.grid()
    plt.show()
```

## Power Functions

### $$ y = x^a  \qquad (x\in R , a \in R)$$

#### $ a = -1   \qquad    y = x^-1  \qquad   x\neq 0$

```python
def f(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        return 1/x

math_function(x,f(x))
    
```

#### $ a = 1/2   \qquad    y = x^{\frac{1}{2}}  \qquad x \geq 0  $

```python
def f(x):

    with np.errstate(divide='ignore', invalid='ignore'):
        return np.sqrt(x)

math_function(x,f(x))
```

#### $ a = 3   \qquad    y = x^2  $

```python
def f(x):

    with np.errstate(divide='ignore', invalid='ignore'):
        return np.power(x,3)

math_function(x,f(x))
```
