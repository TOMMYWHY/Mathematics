# Mathmatics basic

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axisartist.axislines import SubplotZero
import math
import random


x = np.linspace(-5,5,101)

def math_function(x,y):
    fig = plt.figure(figsize=(6,4),dpi=100)
    ax = SubplotZero(fig, 1, 1, 1)
    fig.add_subplot(ax)
    
    ax.axis["xzero"].set_visible(True)
    ax.axis['yzero'].set_visible(True)
    ax.axis["xzero"].set_axisline_style("-|>")


    plt.plot(x,y,color="m")
    plt.grid()
    plt.show()
```

## Power Functions

### $$ y = x^a  \qquad (x\in R , a \in R)$$

#### $ a = -1   \qquad    y = x^-1  \qquad   x\neq 0$

```python
def f(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        return 1/x

math_function(x,f(x))
    
```

#### $ a = 1/2   \qquad    y = x^{\frac{1}{2}}  \qquad x \geq 0  $

```python
def f(x):

    with np.errstate(divide='ignore', invalid='ignore'):
        return np.sqrt(x)

math_function(x,f(x))
```

#### $ a = 3   \qquad    y = x^3  $

```python
def f(x):

    with np.errstate(divide='ignore', invalid='ignore'):
        return np.power(x,3)

math_function(x,f(x))
```

### $$  y = \sqrt[a]{x}   \qquad  ===  \qquad   y = \log_a x  $$

## Trigonometric functions

$y = \sin x$

```python
def f(x):
    return np.sin(x)

math_function(x,f(x))
```

$y = \cos x$

```python
def f(x):
    return np.cos(x)

math_function(x,f(x))
```

$y = \tan x$

```python
def f(x):
    with np.errstate(divide='ignore', invalid='ignore'):
        y = np.tan(x)
        y[:-1][np.diff(y) < 0] = np.nan
        return y

x = np.arange(-2*np.pi, 2*np.pi, 0.1)
plt.plot(x, f(x))
plt.axis([-2*np.pi, 2*np.pi, -2, 2])

plt.grid()
plt.show()

# math_function(x,f(x))
```
